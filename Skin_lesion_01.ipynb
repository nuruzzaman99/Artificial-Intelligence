#Normalize data

```python
x_train = train_derm_img / 255.0
x_test = test_derm_img / 255.0
x_valid = valid_derm_img / 255.0
```

#Define loss function

```python
import tensorflow as tf
import math
```

```python
def my_loss(y_true,y_pred):
    ts = 1.0
    err = tf.abs(y_true - y_pred)
    is_sml_err = err < ts
    sm_err = tf.square(err) / 2.0
    big_err = ts*(err -(0.5 * ts))
    return tf.where(is_sml_err, sm_err, big_err)
```

```python
tf.keras.optimizers.Adam(learning_rate=0.0001, 
                         beta_1=0.9, 
                         beta_2=0.999, 
                         epsilon=1e-07, 
                         amsgrad = False, 
                         name='Adam')
```

#Build the model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import RMSprop
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(400, 400, 3)))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))


model.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

#model.add(Conv2D(256, kernel_size=(3, 3), activation='relu'))
#model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())

model.add(Dense(128, activation='relu'))

model.add(Dense(3, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer= 'Adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train_hc, 
                    epochs=5,
                    steps_per_epoch = 50,
                    batch_size=1, 
                    validation_data = (x_valid,y_valid_hc))
```

#Test the model

```python
result = model.evaluate(x_test, y_test_hc)
```
